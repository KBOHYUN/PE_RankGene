```python
#유전자 한개에 대해서 aml과 all에 대한 ttest
import scipy.stats
from pandas import DataFrame as df

label_name=[]
pvalue=[]
statistic=[]
tu=[] #라벨과 pvalue
tu1=[] #라벨과 statistic
tu2=[] #라벨과 statistic, pvalue


for i in range(len(df_all.columns)):
    lname=df_all.columns[i]
    ttest=scipy.stats.ttest_ind(df_all[lname], df_aml[lname])
    
    if(ttest[0]<0):
        value=abs(ttest[0])
    temp1=(lname,value)
    tu1.append(temp1)
    
    abs(ttest[1]) #절대값
    if(ttest[1]<0.01):
        temp=(lname,ttest[1])
        temp2=(lname,value,ttest[1])
        tu.append(temp)
        tu2.append(temp2)
```
